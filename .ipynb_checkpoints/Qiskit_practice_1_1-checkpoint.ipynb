```python
# 기본적인 quantum state 생성
# state_drawer를 통한 여러 state 상태 출력하기
# display 함수로 여러 state 상태 한꺼번에 출력하기
```

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import state_drawer

qc=QuantumCircuit(2)

ket=Statevector(qc)

display(ket.draw())
display(state_drawer(ket, output='latex'))
display(state_drawer(ket, output='text'))
display(state_drawer(ket, output='paulivec'))
display(state_drawer(ket, output='city'))
display(state_drawer(ket, output='bloch'))
display(state_drawer(ket, output='qsphere'))

#여기서, display함수는 단순히 '보여주는' 기능만 할 뿐, 보여주는 값을 '출력값'으로 지정하지 않음
#따라서 display로 보여지는 값은 cell을 할당받지 않은 것을 볼 수 잇음. 
#하지만 그냥 state_drawer만 쓸 경우, 보여지는 결과값을 '출력값'으로 등록함.
#따라서 그냥 state_drawer만 썼을 경우, 결과가 cell 하나를 할당 받음을 알 수 있음.
#Cell의 출력결과를 output으로 지정해야 할 필요가 있을 경우엔 display를 하면 안됨.
#예를 들어, assert Out[10].data == Out[20].data 와 같이 cell 10의 값과 cell 20의 값이 같은지 비교하고싶은 경우,
#이전 cell [9], cell[19]에서 display로 출력해서는 안됨.
```

```python
# assert 함수를 통해 상태 검증 (주로 디버깅 할때 씀)
# ket이 00 state가 아니면 assertion error가 남.
```

```python
assert ket==Statevector.from_label('00')
```

```python
# Qiskit의 qubit 순서에 유의해야한다.
# Qiskit은 오른쪽부터 왼쪽으로 qubit 순서를 붙인다. 이 방식을 little-endian 이라고 한다.
# 반면, 대부분 이론으로 배울 땐 왼쪽에서 오른쪽으로 qubit 순서를 붙인다. 이 방식을 big-endian 이라고 한다.

# 예를 들어, ket(00) 상태가 있을 때, qc.x(1)을 하게 되면, qiskit에서는 ket(10) 상태를 출력한다. 
# 왜냐하면, Index가 qubit 0이 오른쪽, qubit 1이 왼쪽에 붙기 때문이다. - little endian

# 반면 우리에게 익숙한 이론방식대로 ket(00) 상태에 qc.x(1)을 한다고 생각하면, ket(01) 을 떠올릴 것이다.
# 왜냐하면, Index를 qubit 0이 왼쪽, qubit 1을 오른쪽에 놓기 때문이다. - big endian   
```

```python
qc=QuantumCircuit(2)
qc.h(1)
ket=Statevector(qc)
state_drawer(ket, output='latex')
```

```python
assert ket==Statevector.from_label('+0')
```
